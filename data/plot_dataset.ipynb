```python
import numpy as np
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os

f = './ClutteredMNIST/mnist60x60_sequence1_4dsistortions8x8.npz'
```

```python
data = np.load(f)

train = data['X_train']
test = data['X_test']

print("Train size", train.shape)
print("Test size ", test.shape)
```

```python
y_train = data['y_train']
y_test = data['y_test']
```

```python
image_size = np.sqrt(train[0].shape[-1]).astype(int)
height = image_size 
width = image_size
for i in range(20):
    img = train[i].reshape((height, width))
    plt.imshow(img, cmap='gray', interpolation='none')
    plt.title("Train - " + "  y: " + str(y_train[i]), fontsize=20)
    plt.axis('off')
    plt.show()
    
for i in range(10):
    img = test[i].reshape((height, width))
    plt.imshow(img, cmap='gray', interpolation='none')
    plt.title("Test - " + "  y: " + str(y_test[i]), fontsize=20)
    plt.axis('off')
    plt.show()
    # save as jpg
    # name = os.path.join(".", "TranslatedMNIST", "examples", "TranslatedMNIST_{}.jpg".format(i))
    # plt.savefig(name, dpi=300, bbox_inches='tight', pad_inches=0)
    
    
```
